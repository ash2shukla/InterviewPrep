```python
from random import random
from collections import namedtuple
import numpy as np

Point = namedtuple('Point', ['x', 'y'])

samples = [Point(x=random(), y=random()) for i in range(10000)]

is_inside_curve = lambda point: (point.x**2 + point.y**2)**0.5 <= 1

points_inside_circle = 0

points_mapped = []
for sample in samples:
    if is_inside_curve(sample):
            points_inside_circle += 1
            points_mapped.append(True)
    else:
        points_mapped.append(False)
(points_inside_circle / len(samples))*4
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=[sample.x for sample in samples],
            y=[sample.y for sample in samples],
            c=points_mapped)
```
