#### Description
Given a number n find the least number of squares needed to sum up to the number.

eg. for 13 = 3^2 + 2^2  so 2 numbers

```python
class Solution:
    def sumOfSquares(self, num: int) -> int:
        squares = 0
        while num > 0:
            sqrt = num**0.5
            num -= int(sqrt)**2 
            squares += 1
        return squares
```

```python
from utils import do_test, test

tests = [
    test([296], 5),
    test([2], 2),
    test([1], 1),
    test([5], 2),
]

do_test(tests, Solution.sumOfSquares)
```

#### Complexity
Its near constant, we try plotting for 1000 numbers.

```python
import matplotlib.pyplot as plt
from functools import partial

x = list(range(1000))
y = list(map(partial(Solution.sumOfSquares, None), x))
plt.plot(x, y)
```
